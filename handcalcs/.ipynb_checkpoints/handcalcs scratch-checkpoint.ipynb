```python
with open("../testprint.tex", "r") as file:
    lines = file.readlines()
```

```python
# Attempted scritp to consolidate each individual math environement
# e.g. "align" or "multline" into a single, continuous environemnt.

latex_acc = []
current_environment = ""
new_environment = ""
document_started = False
for line in lines:

    if r"\begin{document}" in line:
        document_started = True
    if document_started == True:
        print("Current Environment: ", current_environment)
        print("New environment: ", new_environment)
        print(line)
        if r"\end{" in line:
            current_environment = line[5:-2]
            if new_environment == current_environment:
                continue
            else:
                latex_acc.append(line)
                continue
        if r"\section" in line:
            latex_acc.append(r"\end{{}}".format(current_environment))
            current_environment = ""
            latex_acc.append(line)
            continue
        if r"\begin{" in line:
            new_environment = line[7:-2].replace("n{", "")
            if new_environment == current_environment:
                continue
            else: 
                latex_acc.append(line + r"\\")
                continue
        latex_acc.append(line)
    else:
        latex_acc.append(line)
print("\n".join(latex_acc))
            
```
